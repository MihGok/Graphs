```python
import networkx as nx,matplotlib.pyplot as plt
```

```python
graph  = nx.DiGraph()
list_nodes = ['а','б','в','г','д','е','ж','з','и','й','к','л']
list_edges = [['а','г'],('а','й'),("б","в"),("в","б"),("в","г"),("в","д"),("г","а"),("г","в"),("д","в"),("е","ж"),("е","з"),("ж","е"),("ж","и"),("ж","й"),("ж","л"),("з","е"),("з","к"),("и","ж"),("и","й"),("й","а"),("й","ж"),("й","и"),("к","з"),("л","ж")]
graph.add_nodes_from(list_nodes)
graph.add_edges_from(list_edges)
nx.draw_circular(graph, with_labels=True)
```

```python
matrix = nx.adjacency_matrix(graph)
print(nx.radius(graph))
print(nx.is_strongly_connected(graph))
```

```python
def create_distance_matrix(cur_graph):
  incidence_matrix = nx.shortest_path_length(cur_graph)
  toOut = True
  for elem in incidence_matrix:
    for elem1 in elem:
      if (isinstance(elem1,dict)):
        print([str(elem1[i]) for i in sorted(elem1.keys())])
```

```python
if (nx.is_strongly_connected(graph)):
  create_distance_matrix(graph)
  print("radius is",nx.radius(graph))
  print("diametr is ", nx.diameter(graph))
  print("center is", nx.center(graph))
```
